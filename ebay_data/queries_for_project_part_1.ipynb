## 1. Find the number of users in the database

## 2. Find the number of users from New York

```python
%%sql
SELECT COUNT(*) 
FROM AuctionUser
WHERE Location = 'New York';
```

## 3.Find the number of auctions belonging to exactly four categories

```python
%%sql 
SELECT COUNT(*)
FROM
(
    SELECT * 
    FROM Category_Belong
    GROUP BY Item_ID
    HAVING COUNT(Item_ID) = 4
);
```

## 4.Find the IDs of auctions with the highest current price

```python
%%sql
SELECT * 
FROM Item
LIMIT 5;
```

```python
%%sql
SELECT Item_ID
FROM Item
WHERE Currently = 
(
    SELECT MAX(Currently)
    FROM Item
);
```


## 5. Find the number of sellers whose rating is higher than 1000.

```python
%sql SELECT * FROM AuctionUser LIMIT 5;
```

```python
%%sql
SELECT Seller_ID
FROM Item
LIMIT 10;
```

```python
%%sql
SELECT COUNT(*)
FROM AuctionUser
WHERE Rating > 1000 AND User_ID IN 
(
    SELECT Seller_ID
    FROM Item
);
```

## 6. Find the number of users who are both sellers and bidders

```python
%%sql
SELECT COUNT(*)
FROM
(
    SELECT User_ID
    FROM Bids
    INTERSECT
    SELECT Seller_ID
    FROM Item
);
```

## 7. Find the number of categories that include at least one item with a bid of more than $100.

```python
%sql SELECT * FROM Bids LIMIT 10;
```

```python
%sql SELECT * FROM Category_Belong LIMIT 5;
```

```python
%%sql
SELECT Category
FROM Category_Belong
WHERE Item_ID IN
(
    SELECT Item_ID
    FROM Bids
    WHERE Amount > 100)
```
